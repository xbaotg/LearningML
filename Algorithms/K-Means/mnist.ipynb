```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

```python
mnist = fetch_openml("mnist_784", as_frame=False)
```

```python
X = mnist.data
y = mnist.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.05)
```

```python
# Visualize
fig, ax = plt.subplots(figsize=(20, 8), nrows=3, ncols=8)

for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i, j].imshow(X[i * ax.shape[0] + j].reshape(28, 28), cmap=plt.cm.gray_r, interpolation="nearest");
```

```python
model = KMeans(n_clusters=10)
```

```python
model.fit(X_train)
```

```python
y_pred = model.predict(X_train)
```

```python
fig, ax = plt.subplots(figsize=(20, 8), nrows=10, ncols=15)

cnt = np.zeros((10,), dtype=np.int8)

for i in range(y_pred.shape[0]):
    v = y_pred[i]
    
    if cnt[v] < ax.shape[1]:
        ax[v, cnt[v]].set_xticklabels([])
        ax[v, cnt[v]].set_yticklabels([])
        ax[v, cnt[v]].imshow(X_train[i].reshape(28, 28), cmap=plt.cm.gray_r)
        cnt[v] += 1
```
