```python
import PIL
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
image = PIL.Image.open("img.jpg")

X = np.array(image)
row_X = X.shape[0]
col_X = X.shape[1]
```

```python
X.shape

# [
#     [[rgb], [rgb], [rgb]],
#     [[rgb], [rgb], [rgb]],
#     [[rgb], [rgb], [rgb]],
#     [[rgb], [rgb], [rgb]],
# ]
```

```python
X = X.reshape(X.shape[0] * X.shape[1], 3)
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
labels = model.predict(X)
```

```python
# [
#     [0, 1, 2, 2, 1],
#     [0, 1, 2, 2, 1],
#     [0, 1, 2, 2, 1],
# ]
```

```python
res = np.zeros((labels.shape[0], 3))
```

```python
res[labels == 0] = np.array([220, 20, 60]) / 255
res[labels == 1] = np.array([0, 0, 0]) / 255
res[labels == 2] = np.array([255, 160, 122]) / 255
```

```python
res = res.reshape(row_X, col_X, 3)
```

```python
plt.imshow(res);
```
