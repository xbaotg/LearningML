```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

X_1 = 1.3 * np.random.rand(20, 1)
y_1 = X_1 + 1.5 * np.random.rand(20, 1)

X_2 = 5 + 1.3 * np.random.rand(20, 1)
y_2 = 1.5 * np.random.rand(20, 1)

X_3 = 2.5 + 1.3 * np.random.rand(20, 1)
y_3 = 4 + 3 * np.random.rand(20, 1)

X_train = np.concatenate((
    np.concatenate((X_1, y_1), axis=1), 
    np.concatenate((X_2, y_2), axis=1), 
    np.concatenate((X_3, y_3), axis=1)), 
    axis=0)

y_train = np.array([i // 20 for i in range(60)])
```

```python
plt.axis([-2, 8, -2, 8])
plt.scatter(X_train[:, 0], X_train[:, 1], c=["r"]*20 + ["g"]*20 + ["b"]*20);
```

```python
X_test = np.concatenate((
    -2 + np.abs(np.random.rand(1000, 1)) * 10,
    -2 + np.abs(np.random.rand(1000, 1)) * 10),
    axis=1
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def predict(n_neighbors, p):
    model = KNeighborsClassifier(n_neighbors=n_neighbors, p=p)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    c = {
        0: 'r',
        1: 'g',
        2: 'b'
    }
    
    colors = [c[i] for i in y_pred]
    plt.scatter(X_train[:, 0], X_train[:, 1], c=["r"]*20 + ["g"]*20 + ["b"]*20);
    plt.scatter(X_test[:, 0], X_test[:, 1], c=colors, alpha=0.1, linewidths=1)
```

```python
predict(4, 2)
```
