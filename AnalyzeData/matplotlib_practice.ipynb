```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("california_cities.csv")
```

```python
df.head()
```

```python
s = df.area_total_sq_mi
p = df.population_total
```

```python
latd, longd = df.latd, df.longd
```

```python
plt.scatter(longd, latd, c=np.log2(p), s=s * 3, edgecolors="black", alpha=0.6);
plt.xlabel("Longd")
plt.ylabel("Latd")
plt.axis("tight")
plt.title("Coordination")
plt.colorbar(label="log$_{2}$ (population)");
```
