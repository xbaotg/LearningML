```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
sy = np.sin(x)
cy = np.cos(x)
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
plt.plot(x, sy, label="sin(x)", linestyle="dashed")
plt.plot(x, cy, label="cos(x)")
plt.axis("equal")

plt.xlabel("x")
plt.ylabel("F(x)")

plt.legend()
```
